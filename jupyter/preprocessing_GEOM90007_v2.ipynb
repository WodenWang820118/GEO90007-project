```python
import pandas as pd
import numpy as np
```

##### The aim of this file is to pre-process dataset including duplicates removal as well as simplification

# Read dataframe with seeating infotmation

```python
industry_types = ['Cafes and Restaurants', 'Pubs, Taverns and Bars', 'Takeaway Food Services']
```

```python
df = pd.read_csv(r"../data/Cafes_and_restaurants__with_seating_capacity .csv",usecols=["Census year", 
                                                                                            "Trading name",
                                                                                             "Industry (ANZSIC4) description",
                                                                                            "Seating type", 
                                                                                            "Number of seats", 
                                                                                            ]) 
df.columns = ["year","name", "industry","seats","num"]
df["year"] = df["year"].fillna(2019)
df_new = df[df['year'] == 2019]
df_new = df_new.fillna(" ")
df_new["seats"] = df_new["seats"].str.slice_replace(0, 8, '')
df_reset=df_new.reset_index(drop=True)

seatInfo= df_reset.loc[df_reset['industry'].isin(industry_types)]
```

# Read raw dataframe without duplicates

```python
raw = pd.read_csv(r"../data/Cafes_and_restaurants__with_seating_capacity .csv",usecols=["Census year", 
                                                                                        "Street address", 
                                                                                        "CLUE small area", 
                                                                                        "Trading name",
                                                                                        "Industry (ANZSIC4) description", 
                                                                                        "x coordinate", 
                                                                                        "y coordinate"]) 
raw.columns = ["year","address", "suburb","name", "industry","longitude","latitude"]
raw["year"] = raw["year"].fillna(2019)
raw = raw.fillna(" ")

raw2019 = raw[raw['year'] == 2019]
raw2019.drop_duplicates(subset="name", inplace = True)
```

```python
raw2019
```

```python
venu = list(raw2019['name'])
add = list(raw2019['address'])
sub = list(raw2019['suburb'])
ind = list(raw2019['industry'])
lon = list(raw2019['longitude'])
lat = list(raw2019['latitude'])
```

```python
location_details = pd.DataFrame({"name": venu, "address": add, "suburb": sub,"industry": ind,"longitude": lon,"latitude": lat})
```

```python
location_details['indoor'] = np.nan
location_details['outdoor'] = np.nan
location_details
```

# Append seat infor after each venue

```python
types = list(seatInfo["seats"])
num = list(seatInfo["num"])
name = list(seatInfo["name"])

name_no_dup = list(location_details["name"])

types[:10], num[:10], name[:10], name_no_dup[:10]
```

```python
in_inx = []
out_inx = []
for i in range(len(types)):
    if types[i] == 'Indoor':
        in_inx.append(i)
    elif types[i] == 'Outdoor':
        out_inx.append(i)
```

```python
for i in range(len(in_inx)):
    venue_name = name[in_inx[i]]
    no_dup_inx = name_no_dup.index(venue_name)
    location_details.loc[no_dup_inx, 'indoor'] = num[in_inx[i]]

for i in range(len(out_inx)):
    venue_name = name[out_inx[i]]
    no_dup_inx = name_no_dup.index(venue_name)
    #print(no_dup_inx, out_inx[i]    )
    location_details.loc[(no_dup_inx), 'outdoor'] = num[out_inx[i]]
```

```python
location_details["indoor"] = location_details["indoor"].fillna(0)
location_details["outdoor"] = location_details["outdoor"].fillna(0)
location_details[:20]
```

```python

location_details = location_details.astype(str)
location_details = location_details.astype({'indoor': int, 'outdoor': int})

location_details =location_details.convert_dtypes()
```
